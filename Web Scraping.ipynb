```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.scrapethissite.com/pages/forms/'
url
```

```python
page = requests.get(url)
page
```

```python
page.text
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
soup
```

```python
print(soup.prettify())
```

```python
columns_header = soup.find_all('th')
columns_header
```

```python
for i in columns_header:
    print(i.text.strip())
```

```python
columns_title = [i.text.strip() for i in columns_header]
columns_title
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = columns_title)
```

```python
df
```

```python
len(df)
```

```python
table= soup.find('table')
table
```

```python
data = table.find_all('tr')
```

```python
data
```

```python
data[1:]
```

```python
for row in data[1:]:
    row_data = row.find_all('td')
    ind_row_data = [x.text.strip() for x in row_data]
    length = len(df)
    df.loc[length] =ind_row_data
```

```python
df
```

```python
len(df)
```

```python
df.to_csv('Hockey_Team1.csv', index = False)
```

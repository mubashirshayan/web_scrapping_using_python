```python
# Import Libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
url = requests.get('https://www.scrapethissite.com/pages/forms/?per_page=1000')
```

```python
soup = BeautifulSoup(url.text,"html")
soup
```

```python
print(soup.prettify())
```

```python
column_header = soup.find_all('th')
column_header
```

```python
column_title =[i.text.strip() for i in column_header]
column_title
```

```python
df =pd.DataFrame(columns=column_title)
```

```python
df
```

```python
table = soup.find("table")
```

```python
table
```

```python
data = table.find_all("tr")
```

```python
for row in data[1:]:
    row_data = row.find_all('td')
    ind_row_data = [x.text.strip() for x in row_data]
    length = len(df)
    df.loc[length] =ind_row_data
```

```python
df
```

```python
df.to_csv("Hockey_Team_Data.csv",index =False)
```
